## Data Processing

```python
ids = pd.read_csv('../new_ids.csv')
```

```python
#Changing to new id

ntrain = train.merge(ids, left_on='msno', right_on='msno', how='inner').drop('msno', axis = 1)
ntransactions = transactions.merge(ids, left_on='msno', right_on='msno', how='inner').drop('msno', axis = 1)
nmembers = members.merge(ids, left_on='msno', right_on='msno', how='inner').drop('msno', axis = 1)
nuser_logs = user_logs.merge(ids, left_on='msno', right_on='msno', how='inner').drop('msno', axis = 1)

print ('Lengths:\n train: %s, ntrain: %s' %(len(train), len(ntrain)))
print (' ntransactions: %s, transactions: %s' %(len(ntransactions), len(transactions)))
print (' nmembers: %s, members: %s' %(len(nmembers), len(members)))
print (' nuser_logs: %s, user_logs: %s' %(len(nuser_logs), len(user_logs)))
```

>OBS: Some transactions / members / user logs seem to have users not in the train / test datasets

```python
ntrain_memb = ntrain.merge(nmembers, on = 'new_id', how='inner')
sns.countplot('is_churn',hue = 'gender', data = ntrain_memb)
```

```python
plt.figure(num=None, figsize=(16, 4), dpi=80)
#plt.title('log scale of daily seconds (censored)');

plt.subplot(2,2,1)
sns.violinplot(user_logs[user_logs['num_25'] < 50]['num_25'])
plt.title('25%')

plt.subplot(2,2,2)
sns.violinplot(user_logs[user_logs['num_50'] < 50]['num_50'])
plt.title('50%')

plt.subplot(2,2,3)
sns.violinplot(user_logs[user_logs['num_985'] < 50]['num_985'])
plt.title('98.5%')

plt.subplot(2,2,4)
sns.violinplot(user_logs[user_logs['num_100'] < 100]['num_100'])
plt.title('100%')
```

```python
plt.figure(num=None, figsize=(16, 4), dpi=80)

plt.subplot(1,4,1)
ntuser_log = ntrain.merge(nuser_logs, on = 'new_id', how='inner')
sns.boxplot(ntuser_log['is_churn'], ntuser_log[ntuser_log['num_25'] < 50]['num_25'])

plt.subplot(1,4,2)
ntuser_log = ntrain.merge(nuser_logs, on = 'new_id', how='inner')
sns.boxplot(ntuser_log['is_churn'], ntuser_log[ntuser_log['num_50'] < 50]['num_50'])

plt.subplot(1,4,3)
ntuser_log = ntrain.merge(nuser_logs, on = 'new_id', how='inner')
sns.boxplot(ntuser_log['is_churn'], ntuser_log[ntuser_log['num_985'] < 50]['num_985'])

plt.subplot(1,4,4)
ntuser_log = ntrain.merge(nuser_logs, on = 'new_id', how='inner')
sns.boxplot(ntuser_log['is_churn'], ntuser_log[ntuser_log['num_100'] < 200]['num_100'])
```

>There are only minor differences in song listening from a churn perspective (although this is only a portion of the dataset)

```python
plt.figure(num=None, figsize=(16, 4), dpi=80)
plt.subplot(1,2,1)
sns.distplot(ntuser_log['num_unq'])
plt.subplot(1,2,2)
sns.boxplot(ntuser_log['is_churn'], ntuser_log[ntuser_log['num_unq']<200]['num_unq'])
```

>Again, nothing clear about a differentiation on churn vs. non-churn here. Perhaps the best way to understand is by a user behavior matrix across time. Some very extreme values of songs listened in one day.

## Ideas for Modeling Data Structure

| User_id | days_active | subscription_plan | churned_cnt | avg_seconds | last_paymt | mode_pymt_id |...|
|---|---|---|---|---|---|---|---|
| 1 | 10| x | 2 |150| 30| 41|...|
| 2 |140| z | 0 |500|  1| 40|...|
| 3 |790| x | 1 |900| 20| 30|...|

From transactions:
- days_registered (number of days registered)
- current_subscription_plan (how do we figure out which current subscription they have?)
- subscription_switch_flag (did they switch annual to monthly, etc.)
- pmt_method_switch_flag (switched payment methods)
- auto_renew_switch_flag (auto-renew to non auto-renew)
- churned_cnt (number of 30 day gaps?)
- last_pmt (days since last payment)

From user logs:
- avg_songs_played (average number of songs played per day)
- avg_length_played (average length of songs played)
- per_days_active (percentage of days that were active i.e. played songs, out of number of days registered)
- per_songs_25 (percentage of songs played under 25 percent)
- per_songs_50 (percentage of songs played under 50 percent)
- per_songs_75 (percentage of songs played under 98.5 percent)
- per_songs_985 (percentage of songs played 100 percent)
    can maybe generate a score of how much of the song is usually played? (sum of % of songs played x% * x%)
- some type of score to indicate decline in activity?
